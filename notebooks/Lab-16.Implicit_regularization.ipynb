## Linear regression


Let's first create a function which generates data. You've already seen this in many previous labs.

````python
def datagen(d, points, m, M, w, sigma):
    X = np.zeros((points,d))
    for i in range(points):
        X[i,:] = np.random.uniform(m, M, d)
    eps = np.random.normal(0, sigma, points)
    y = np.dot(X,w) + eps 
    return X, y
````

#### Defining the auxiliary functions

````python
# functions

def get_accuracy(logit, target):
    corrects = (torch.max(logit, 1)[1].view(target.size()).data == target.data).sum()
    accuracy = 100.0 * corrects / target.size(0)
    return accuracy.item()


def compute_weight_norm(model):
    norm = 0.0
    for name, param in model.named_parameters():
      if "weight" in name:
        norm += torch.norm(param.data, p= 2)
    return norm.cpu().item()


def train_model(model, num_epochs, trainloader, criterion, optimizer):

    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

    model = model.to(device)

    losses = []
    accs = []

    norms = []
    
    for epoch in range(num_epochs):
        train_running_loss = 0.0
        train_acc = 0.0

        # Set the model to training mode
        model = model.train()
        start = time()
        ## training step
        for i, (images, labels) in enumerate(trainloader):
            
            images = images.to(device)
            labels = labels.to(device)

            ## forward + backprop + loss
            logits = model(images)
            loss = criterion(logits, labels)

            # Reset the gradients to zero
            optimizer.zero_grad()

            loss.backward()

            ## update model params
            optimizer.step()

            train_running_loss += loss.item()
            train_acc += get_accuracy(logits, labels)
        
            
        losses.append(train_running_loss / i)
        accs.append(train_acc/i)
        
        norms.append(compute_weight_norm(model))
        model.eval()
        print(f"Epoch: {epoch+1} | Loss: {train_running_loss / i:.4f} | Train Accuracy: {train_acc/i:.4f} | Time: {time()-start:.2f}") 
    
    return losses, accs, norms
````

## YOLOv8 Object Detection on Real Images


So far we built a CNN classifier from scratch. Now we explore YOLOv8, a state-of-the-art model for object detection — a harder task where the model must simultaneously locate and classify multiple objects in an image, outputting bounding boxes alongside class labels.
YOLOv8 is pretrained on COCO (80 classes, ~330k images). We will run it on a real photograph and visualise its predictions.

````python
!pip install ultralytics --quiet

import torch
from ultralytics import YOLO
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import requests
from io import BytesIO

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

model_yolo = YOLO("yolov8n.pt")  # Downloads weights automatically

print("Classes this model knows:")
print(list(model_yolo.names.values()))
````

## Run detection on real images

````python
def load_image_from_url(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content)).convert("RGB")

# A busy street scene with pedestrians, vehicles, etc.
#IMAGE_URL = "https://ultralytics.com/images/bus.jpg"
IMAGE_URL = "https://external-content.duckduckgo.com/iu/?u=https%3A%2F%2Fc8.alamy.com%2Fcomp%2FETDCH8%2Findia-indian-mumbai-dadar-west-rush-hour-people-commuters-pedestrians-ETDCH8.jpg&f=1&nofb=1&ipt=4d3f2790511b8b794e7965db09466d1f29e723f670ceb9fa388972bc41a49138"
image = load_image_from_url(IMAGE_URL)

results = model_yolo(image, verbose=False)
result  = results[0]

print(f"Detected {len(result.boxes)} objects.")
for box, conf, cls_id in zip(
    result.boxes.xyxy.cpu().numpy(),
    result.boxes.conf.cpu().numpy(),
    result.boxes.cls.cpu().numpy().astype(int)
):
    print(f"  {result.names[cls_id]:<15} confidence: {conf:.2f}  box: {np.round(box).astype(int).tolist()}")
````

## Visualise Predictions

````python
def visualise_detections(image, result, conf_threshold=0.3):
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))

    # Left: original image
    axes[0].imshow(image)
    axes[0].set_title("Original image")
    axes[0].axis("off")

    # Right: detections
    axes[1].imshow(image)
    axes[1].set_title(f"YOLOv8n detections (conf ≥ {conf_threshold})")
    axes[1].axis("off")

    # Assign a consistent colour to each class
    cmap = plt.cm.get_cmap("tab20", len(result.names))
    
    for box, conf, cls_id in zip(
        result.boxes.xyxy.cpu().numpy(),
        result.boxes.conf.cpu().numpy(),
        result.boxes.cls.cpu().numpy().astype(int)
    ):
        if conf < conf_threshold:
            continue
        x1, y1, x2, y2 = box
        color = cmap(cls_id)

        rect = patches.Rectangle(
            (x1, y1), x2 - x1, y2 - y1,
            linewidth=2, edgecolor=color, facecolor="none"
        )
        axes[1].add_patch(rect)
        axes[1].text(
            x1, y1 - 6,
            f"{result.names[cls_id]} {conf:.2f}",
            color="white", fontsize=9,
            bbox=dict(facecolor=color, alpha=0.8, pad=2, edgecolor="none")
        )

    plt.tight_layout()
    plt.show()

visualise_detections(image, result)
````
